```python
from scipy.spatial.distance import cosine
```

```python
#using same tokenizer and model
```

```python
def predict(sentence):
  encoded_input=tokenizer(sentence,return_tensors="pt")
  model_output=model(**encoded_input)[0]
  return model_output
```

```python
#setting up sentences
```

```python
# Defining the sentences
sentence1 = "There was a fly drinking from my soup"
sentence2 = "There is a fly swimming in my juice"
# sentence2 = "To become a commercial pilot, he had to fly for 1500 hours." # second fly example

# Tokenizing the sentences
tokens1 = tokenizer.tokenize(sentence1)
tokens2 = tokenizer.tokenize(sentence2)
```

```python
# Getting model predictions for the sentences
out1 = predict(sentence1)
out2 = predict(sentence2)
```

```python
out1
```

```python
out2
```

```python
emb1 = out1[0:, tokens1.index("fly"), :].detach()[0]
emb2 = out2[0:, tokens2.index("fly"), :].detach()[0]

# emb1 = out1[0:, 3, :].detach()
# emb2 = out2[0:, 3, :].detach()
```

```python
cosine(emb1,emb2)
```

```python
#sentence similarity
```

```python
import torch
```

```python
def get_sentence_embedding(sentence):
    tokens = tokenizer(sentence, return_tensors='pt', truncation=True, padding=True)
    with torch.no_grad():
        output = model(**tokens)

    cls_embedding = output.last_hidden_state[:, 0, :]  # Shape: (1, hidden_size)
    return cls_embedding.squeeze(0).numpy()
```

```python
# Sentences to compare
sentence1 = "The quick brown fox jumps over the lazy dog."
sentence2 = "A fast, brown animal leaps over a sleepy canine."

# Get embeddings
embedding1 = get_sentence_embedding(sentence1)
embedding2 = get_sentence_embedding(sentence2)
```

```python
cosine(embedding1,embedding2)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity([embedding1], [embedding2])
print(f"Cosine Similarity: {similarity[0][0]}")
```

#MASKED TOKENS

```python
import numpy as np
from scipy.special import softmax
from transformers import AutoModelForMaskedLM, AutoTokenizer
```

```python
model = AutoModelForMaskedLM.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
mask=tokenizer.mask_token #defining a mask token for prediction
```

```python
mask
```

```python
sentence= f"I want to {mask} me"
```

```python
tokens=tokenizer.tokenize(sentence)
```

```python
tokens
```

```python
tokenize_input=tokenizer(sentence, return_tensors="pt")
output=model(**tokenize_input)
output
```

```python
#logits: logit is for probability score of tokens that could be there
```

```python

# Detaching the logits from the model output and converting to numpy array
logits = output.logits.detach().numpy()[0]
```

```python
logits
```

```python
masked_logits=logits[tokens.index(mask)+1]
confidence_scores=softmax(masked_logits)
```

```python
confidence_scores
```

```python
for i in np.argsort(confidence_scores)[::-1][:5]:
    pred_token = tokenizer.decode(i)
    score = confidence_scores[i]

    # print(pred_token, score)
    print(sentence.replace(mask, pred_token))
```
