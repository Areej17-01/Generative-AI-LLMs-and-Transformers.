```python

from transformers import pipeline
import torch
```

```python
pipe = pipeline("text-generation", model="unsloth/Llama-3.2-1B-Instruct", torch_dtype=torch.bfloat16, device_map="auto")
```

```python

messages = [
    {"role":"system", "content":  "You are a highly logical AI. Think step by step and explain your reasoning clearly."},
    {"role": "user", "content": "A 5-year-old boy presents with bleeding gums, malnutrition, and small, curled hairs, while having a normal CBC except for mild anemia with an MCV of 85. His diet primarily consists of boiled and preserved vegetables due to financial hardship. What nutritional deficiency is most likely responsible for his symptoms?"},
]
```

```python
prompt = pipe.tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
```

```python


outputs = pipe(prompt, max_new_tokens=128)

print(outputs[0]["generated_text"])
```

```python
import yaml

train_config = {
    "base_model": "unsloth/Llama-3.2-1B-Instruct",

    "datasets": [{"path": "Yukang/LongAlpaca-12k", "type": "alpaca"}],
    "output_dir": "./models/LongAlpaca",

    # model params
    "sequence_length": 1024,
    "bf16": "auto",
    "tf32": False,

    # training params
    "micro_batch_size": 1,
    "num_epochs": 1,
    "optimizer": "adamw_bnb_8bit",
    "learning_rate": 0.0002,

    "logging_steps": 1,

    # LoRA / qLoRA
    "adapter": "qlora",
    "lora_r": 32,
    "lora_alpha": 16,
    "lora_dropout": 0.05,
    "lora_target_linear": True,

    # Gradient Accumulation
    "gradient_accumulation_steps": 1,

    # Gradient Checkpointing
    "gradient_checkpointing": True,

    # Low Precision
    "load_in_8bit": False,
    "load_in_4bit": True,

    # Flash Attention
    "flash_attention": False,
}





# Write the YAML file
with open("specialised_train.yml", 'w') as file:
    yaml.dump(train_config, file)
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig
import torch
#AreejMehboob17/Llama-3.2-1B-LongAlpaca
model_path = "./models/LongAlpaca/merged"

# Load the tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_path, legacy=False)

# Enable 8-bit loading to reduce VRAM usage
quantization_config = BitsAndBytesConfig(load_in_8bit=True)

# Load the model in 8-bit
model = AutoModelForCausalLM.from_pretrained(
    model_path,
    device_map="auto",
    quantization_config=quantization_config
)

# Initialize the text-generation pipeline
pipe = pipeline("text-generation", model=model, tokenizer=tokenizer)

messages = [
    {"role":"system", "content": query+"<END>"},
]

prompt = pipe.tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)

output = pipe(prompt)
print("Model Output:", output[0]['generated_text'])
```
