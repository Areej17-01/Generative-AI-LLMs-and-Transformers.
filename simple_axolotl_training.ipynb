```python
import yaml


yaml_string = """
# model params
base_model: TinyLlama/TinyLlama-1.1B-Chat-v1.0
model_type: LlamaForCausalLM
tokenizer_type: LlamaTokenizer


# dataset params
datasets:
  - path: jaydenccc/AI_Storyteller_Dataset
    type:
      system_prompt: ""
      field_system: system
      field_instruction: synopsis
      field_output: short_story
      format: "<|user|>\n {instruction} </s>\n<|assistant|>"
      no_input_format: "<|user|> {instruction} </s>\n<|assistant|>"

output_dir: ./models/TinyLlama_Storyteller

# model params
sequence_length: 1024
bf16: auto
tf32: false

# training params
batch_size: 2
micro_batch_size: 4
num_epochs: 2
optimizer: adamw_bnb_8bit
learning_rate: 0.0001
lr_scheduler_type: cosine
warmup_steps: 100




logging_steps: 1
"""
```

```python
yaml_dict = yaml.safe_load(yaml_string)
yaml_dict
```

```python
# Specify your file path
file_path = 'test_axolotl.yaml'

# Write the YAML file
with open(file_path, 'w') as file:
    yaml.dump(yaml_dict, file)
```
